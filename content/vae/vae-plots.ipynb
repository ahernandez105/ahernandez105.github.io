```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import math
import os

plt.rcParams["font.family"] = "serif"
plt.rcParams['font.sans-serif'] = ['DejaVu']
plt.rcParams["font.size"] = 70
plt.style.use('seaborn-dark-palette')
path = '/Users/angelhernandez/GitHub/ach_blog/content/images/vae'
```

```python
mus = [0,3]
variances = [1,1]
sigmas = [math.sqrt(variances[0]),math.sqrt(variances[1])]
Xs = [np.linspace(mus[x] - 3*sigmas[x],mus[x] + 3*sigmas[x],100) for x in range(2)]
Ys = [stats.norm.pdf(Xs[x],mus[x],sigmas[x]) for x in range(2)]
annotates = [
    r'$p_{\theta}(\mathbf{z|x})$',
    r'$q_{\phi}(\mathbf{z|x})$'
]
fig, (ax1, ax2)  = plt.subplots(nrows=1,ncols=2,figsize = (30,15))

# subplot 1
ax1.set_title("KL Divergence " + r"$>$ " + str(0))
ax1.plot(Xs[0], Ys[0],color="black")
ax1.plot(Xs[1],Ys[1],color="black")
ax1.fill_between(Xs[0],Ys[0],alpha=0.8)
ax1.fill_between(Xs[1],Ys[1],alpha=0.8)
ax1.annotate(annotates[0],(-2.75,0.25),fontsize=70)
ax1.annotate(annotates[1],(4.25,0.25),fontsize=70)
ax1.set_ylabel("Density")
ax1.grid(True,alpha=0.3,linestyle="--")

# subplot 2
ax2.set_title("KL Divergence " + r"$=$ " + str(0))
ax2.plot(Xs[0], Ys[0],color="black")
ax2.plot(Xs[0],Ys[0],color="black")
ax2.fill_between(Xs[0],Ys[0],alpha=0.8)
ax2.fill_between(Xs[0],Ys[0],alpha=0.5)
ax2.annotate(annotates[0],(-2.75,0.25),fontsize=70)
ax2.annotate(annotates[1],(1.75,0.25),fontsize=70)
ax2.grid(True,alpha=0.3,linestyle="--")
fig.set_size_inches(80,40)
plt.close('all')

fig.savefig(f'{path}/kl-divergence.png',bbox_inches='tight')
fig
```
